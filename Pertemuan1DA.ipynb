```python
asia_country = ["Brunei", "Cambodia", "Indonesia", "Laos", "Malaysia", "Myanmar", "Philippines", "Singapore", "Thailand", "Timor-Leste", "Vietnam", "China", "Japan", "South Korea", "North Korea", "Mongolia", "Taiwan", 
                "India", "Pakistan", "Bangladesh", "Nepal", "Sri Lanka", "Bhutan", "Maldives", "Turkey", "Iran", "Iraq", "Saudi Arabia", "Jordan", "Syria", "Lebanon", "Israel", "Palestine", "United Arab Emirates", "Qatar", "Kuwait", "Oman", "Yemen", "Kazakhstan", "Uzbekistan", "Turkmenistan", "Kyrgyzstan", "Tajikistan"]

df_asia_country = df[df["Country or region"].isin(asia_country)]
```

```python
asia_country_without_indonesia = ["Brunei", "Cambodia", "Laos", "Malaysia", "Myanmar", "Philippines", "Singapore", "Thailand", "Timor-Leste", "Vietnam", "China", "Japan", "South Korea", "North Korea", "Mongolia", "Taiwan", 
                "India", "Pakistan", "Bangladesh", "Nepal", "Sri Lanka", "Bhutan", "Maldives", "Turkey", "Iran", "Iraq", "Saudi Arabia", "Jordan", "Syria", "Lebanon", "Israel", "Palestine", "United Arab Emirates", "Qatar", "Kuwait", "Oman", "Yemen", "Kazakhstan", "Uzbekistan", "Turkmenistan", "Kyrgyzstan", "Tajikistan"]

asia_country = asia_country_without_indonesia + ["Indonesia"]

# Membuat DataFrame baru hanya dengan negara-negara Asia yang sudah disesuaikan
df_asia_with_indonesia = df[df["Country or region"].isin(asia_country)]
```

Posisi Indonesia Ke-22 di ASIA

```python
#Asia
plt.figure(figsize=(20, 10))
ax = sns.barplot(x='Country or region', y='Score', data=df_asia_with_indonesia, palette='Blues')
plt.title('Score Kebahagiaan di Setiap Negara Asia dengan Indonesia')
plt.xlabel('Negara di Asia')
plt.ylabel('Score')
plt.xticks(rotation=45)  # Mengatur rotasi label sumbu x agar lebih mudah dibaca

# Mendapatkan batas-batas sumbu x dan y
xmin, xmax = ax.get_xlim()
ymin, ymax = ax.get_ylim()

# Tambahkan garis putus-putus untuk menandakan skor kebahagiaan rata-rata (Global Mean)
plt.axhline(mean_score, linestyle='--', color='red', label='Global Mean')

# Tambahkan teks "Global Mean" di luar batas sumbu x
ax.text(xmax + 0.1, mean_score, f'Global Mean: {mean_score:.2f}', ha='left', va='center', fontsize=12, color='blue')

# Tambahkan teks untuk setiap bar
for index, value in enumerate(df_asia_with_indonesia['Score']):
    plt.text(index, value + 0.1, f'{value:.5f}', ha='center', va='bottom', fontsize=11)

plt.show()
```

```python
indjpn_country =  ["Indonesia", "Japan"]

# Columns to include in the radar chart
columns = ['GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']

# Number of variables we're plotting.
num_vars = len(columns)

# Compute angle each bar is centered on:
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop by appending the first angle at the end.

# Draw the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for k, (idx, row) in enumerate(df_indjpn_country.iterrows()):
    data = row[columns].tolist() + row[columns].tolist()[:1]  # Complete the loop
    ax.fill(angles, data, alpha=0.4, label=row['Country or region'], color=colors_div[3*k], edgecolor=colors_div[3*k])
    ax.plot(angles, data)

# Labels for each point
ax.set_xticks(angles[:-1])
col_name = ['GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']
ax.set_xticklabels(col_name)

# Title and legend
plt.title('Perbandingan Faktor Happines Indonesia dengan Jepang', size=15, color=colors_red[1], y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```

```python
fs_country =  ["Finland", "South Sudan"]

df_fs_country = df[df["Country or region"].isin(fs_country)]
```

```python
fs_country =  ["Finland", "South Sudan"]

df_fs = df_fs_country[df_fs_country["Country or region"].isin(fs_country)]
for i , r in df_fs.iterrows():
  print(r)
```

```python
fs_country = ["Finland", "South Sudan"]
df_fs_country = df[df["Country or region"].isin(fs_country)]

# Pastikan DataFrame berisi data yang sesuai
print(df_fs_country)

# Columns to include in the stacked bar chart
columns = ['GDP per capita',
           'Social support', 'Healthy life expectancy',
           'Freedom to make life choices',
           'Generosity', 'Perceptions of corruption']

# Set data for each country
category_finland = df_fs_country[df_fs_country['Country or region'] == 'Finland'][columns].values.flatten()
category_sudanselatan = df_fs_country[df_fs_country['Country or region'] == 'South Sudan'][columns].values.flatten()

# Set the labels for each category
labels = columns

# Set the positions for the bars
x = np.arange(len(labels))

# Set the width of the bars
width = 0.35

# Plot the stacked bars
fig, ax = plt.subplots(figsize=(12, 8))

# Plot bars for Finland
bars_finland = ax.barh(x - width/2, category_finland, width, label='Finland', color='red')

# Plot bars for South Sudan on top of Finland's bars
bars_sudanselatan = ax.barh(x + width/2, category_sudanselatan, width, label='South Sudan', color='blue')

# Add labels to the bars
for i, (category, value) in enumerate(zip(labels, category_finland)):
    ax.text(value, i - width/2, f'{value:.2f}', ha='left', va='center', color='black')

# Add labels to the bars
for i, (category, value) in enumerate(zip(labels, category_sudanselatan)):
    ax.text(value, i + width/2, f'{value:.2f}', ha='left', va='center', color='black')

# Add labels, title, and legend
ax.set_xlabel('Explained by Score')
ax.set_ylabel('Factors')
ax.set_title('Comparison of Happiness Factors: Finland and South Sudan')
ax.set_yticks(x)
ax.set_yticklabels(labels)
ax.legend()

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(left=0.6, right=1, top=0.9, bottom=0.6)

# Show plot
plt.show()
```
